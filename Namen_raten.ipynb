# LLM-Systemprompt erstellen

# Task
Generieren Sie einen Systemprompt f√ºr das LLM, der die vorbereiteten Mitarbeiterdaten enth√§lt, um dem LLM den notwendigen Kontext f√ºr die Erkennung zu geben.

# ü§ñ Gemini API Anfrage: Mitarbeiter-Identifikation

Diese Zelle f√ºhrt die Kommunikation mit der Google Gemini API durch.

**Funktionsweise:**
1.  **Authentifizierung:** Initialisiert den `genai.Client` mit dem API-Key.
2.  **Modellwahl:** Fragt interaktiv ab, ob das leistungsst√§rkere `Pro`-Modell oder das schnellere `Flash`-Modell verwendet werden soll.
3.  **Generierung:** Sendet den `system_prompt` und die Suchanfrage (`user_combined_input`) an die KI.
4.  **Fehlerbehandlung:** F√§ngt spezifische API-Fehler (z.B. *Quota √ºberschritten* oder *Modell nicht gefunden*) ab und gibt detaillierte Diagnosedaten aus.

> **Hinweis:** Stellen Sie sicher, dass die Variablen `user_combined_input` und `system_prompt` in einer vorherigen Zelle definiert wurden.

```python
import google.genai as genai
import os
from google.api_core.exceptions import TooManyRequests, NotFound
from google.colab import userdata

# Configure API key
api_key = userdata.get('GOOGLE_API_KEY')
#api_key = "Hier k√∂nnte ihr GOOGLE API Schl√ºssel stehen" # Re-using the existing api_key from the kernel state.
if not api_key:
    api_key = input("Bitte geben Sie Ihren Google API-Schl√ºssel ein: ")

# Initialize the generative AI client
client = genai.Client(api_key=api_key)

print("Bitte w√§hlen Sie ein Modell aus: ")
chosen_model_input = input("1 f√ºr 'gemini-pro-latest', 2 f√ºr 'gemini-flash-latest' (Standard ist 2): ").strip()

if chosen_model_input == '1':
    model_name = 'gemini-pro-latest'
elif chosen_model_input == '2' or chosen_model_input == '': # Default to 'gemini-flash-latest' if '2' or empty
    model_name = 'gemini-flash-latest'
else:
    print(f"Ung√ºltige Modellauswahl '{chosen_model_input}'. Verwende Standardmodell 'gemini-flash-latest'.")
    model_name = 'gemini-flash-latest'

print(f"Ausgew√§hltes Modell: {model_name}")

# Create user message from inputs (user_combined_input should be available from previous cell)
user_message = (
    f"Die Person, die ich suche, hat Hobbys und Eigenschaften, die sich wie folgt beschreiben lassen: '{user_combined_input}'. " +
    f"Welcher Mitarbeiter aus der Liste passt am besten zu dieser Beschreibung? " +
    f"Antworte nur mit dem Namen des Mitarbeiters."
)

try:
    # Send prompts to LLM and get response directly using client.models.generate_content
    response = client.models.generate_content(model=model_name, contents=[system_prompt, user_message])

    # Check if there's any text in the response
    if response.candidates:
        #The response.text attribute typically holds the generated text content from the first candidate of the LLM's response.
        identified_employee = response.text.strip()
        print(f"Der identifizierte Mitarbeiter ist: {identified_employee}")
        print(f"Der Anfragetext ist: {user_combined_input}")

    else:
        print("Die KI konnte keinen Mitarbeiter identifizieren oder die Antwort wurde blockiert.")
        if response.prompt_feedback and response.prompt_feedback.block_reason:
            print(f"Blockierungsgrund: {response.prompt_feedback.block_reason}")
        if response.candidates and response.candidates[0].finish_reason:
            print(f"Finish Reason des Kandidaten: {response.candidates[0].finish_reason}")
        if response.candidates and response.candidates[0].safety_ratings:
            print("Sicherheitsbewertungen:")
            for rating in response.candidates[0].safety_ratings:
                print(f"  {rating.category}: {rating.probability}")

except TooManyRequests as e:
    print(f"Fehler: Kontingent √ºberschritten (TooManyRequests). Ihre Anfrage konnte nicht verarbeitet werden.")
    print(f"M√∂gliche Gr√ºnde:\n- Sie haben Ihr aktuelles Nutzungslimit erreicht.")
    print(f"Bitte √ºberpr√ºfen Sie Ihr Kontingent auf der Google Cloud Console oder AI Studio.\nVersuchen Sie es sp√§ter erneut.")
    print(f"Details: {e}")
except NotFound as e:
    print(f"Fehler: Modell nicht gefunden (NotFound). Das angegebene Modell '{model_name}' ist m√∂glicherweise nicht verf√ºgbar.")
    print(f"M√∂gliche Gr√ºnde:\n- Der Modellname ist falsch geschrieben oder existiert nicht mehr.\n- Das Modell ist in Ihrer Region nicht verf√ºgbar.")
    print(f"Bitte √ºberpr√ºfen Sie die Liste der verf√ºgbaren Modelle und passen Sie 'model_name' entsprechend an.")
    print(f"Details: {e}")
except Exception as e:
    print(f"Ein unerwarteter Fehler ist aufgetreten: {e}")
```
